Este documento tem por objetivo principal descrever, principalmente para pessoas que não tenham conhecimento algum na área, todos os procedimentos necessários para a instalação e utilização da linguagem Pyhton, da linguagem Julia e do controlador de versão Git. Com estas ferramentas é possível criar diversos programas e salvá-los online, através de um controlador de versão. Este conjunto é essencial para quem deseja manter seus trabalhos atualizados e organizados, seja um programador profissional ou um estudante.

Antes de começar é importante conhecer um pouco de cada ferramenta citada e de outras que surgiram ao longo do processo. Para se chegar ao resultado final, a geração do presente documento, e de outros com foco nas linguagens Python e Julia, e sua disponibilização no GitHub, através de uma url, instalou-se inicialmente Anaconda, uma distribuição Python completamente gratuita. 

Python é uma linguagem de programação criada por Guido van Rossum em 1991. Os objetivos do projeto da linguagem eram produtividade e legibilidade. Em outras palavras, Python é uma linguagem que foi criada para produzir código bom e fácil de manter de maneira rápida. Entre as características da linguagem que ressaltam esses objetivos estão: baixo uso de caracteres especiais, o uso de identação para marcar blocos, quase nenhum uso de palavras-chave voltadas para a compilação, entre outras.

Outra liguagem utilizada neste projeto foi a Julia, que começou a ser desenvolvida em 2009, tendo uma versão de código aberto divulgado em fevereiro de 2012. Julia é uma linguagem de programação dinâmica de alto nível projetada para atender os requisitos da computação de alto desempenho numérico e científico, sendo também eficaz para a programação de propósito geral. Julia é escrita em C, C++, e Scheme, usando a estrutura do compilador LLVM, enquanto a maior parte da biblioteca padrão de Julia é implementada na própria Julia.

Outro aplicativo utilizado foi o Jupyter Notebook, no qual este documento foi gerado. O Jupyter Notebook é uma aplicação web que permite criar e compartilhar documentos que contêm equações, visualizações e texto explicativo. Os usos incluem: limpeza de dados e transformação, simulação numérica, modelagem estatística, aprendizado de máquina e muito mais. O Notebook possui suporte para mais de 40 linguagens de programação, incluindo Python e Julia, as que foram utilizadas aqui.

Na última parte deste processo foi instalado o Git, que é um sistema de controle de versão livre e de código aberto distribuído projetado para lidar com tudo, desde pequenos a grandes projetos com rapidez e eficiência. Através dele é possível desenvolver projetos na qual diversas pessoas podem contribuir simultaneamente no mesmo, editando e criando novos arquivos e permitindo que os mesmos possam existir sem o risco de suas alterações serem sobrescritas.

Para hospedar os projetos criados com o Git utiliza-se o Github, que é um serviço web que oferece diversas funcionalidades extras aplicadas ao Git. Além disso, quase todos os projetos/frameworks/bibliotecas sobre desenvolvimento open source estão no github, e é possível acompanhá-los através de novas versões, contribuir informando bugs ou até mesmo enviando código e correções.
